```python
from scripts.accounting_map import map_to_accounting
import pandas as pd
df = pd.read_csv("dataset_rf.csv")

entries = []
for idx, row in df.iterrows():
    chosen_label = row.get('gemini_label', None)
    if not chosen_label or chosen_label == "ERROR":
        chosen_label = row['rf_label']
    entry = map_to_accounting(chosen_label, row['text'], total=row.get('total', None))
    entries.append({"file": row['file'], "chosen_label": chosen_label, **entry})

entries_df = pd.DataFrame(entries)
entries_df.to_csv("results.csv", index=False)
```
